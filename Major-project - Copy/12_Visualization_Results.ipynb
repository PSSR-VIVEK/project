Generate publication-ready visualizations for the project.

```
!pip -q install pandas matplotlib scikit-learn

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from google.colab import files

print("Upload training_dataset.csv")
u=files.upload()
train=pd.read_csv(list(u.keys())[0])

print("Upload cnn_features.csv")
u=files.upload()
cnn=pd.read_csv(list(u.keys())[0])

print("Upload transformer_features.csv")
u=files.upload()
tr=pd.read_csv(list(u.keys())[0])

# -------------------------
# 1. Dataset Distribution
# -------------------------
labels=train["event_id"].astype(str).apply(lambda x:"FLASH" if "FLASH" in x.upper() else "HEAT")
counts=labels.value_counts()

plt.figure(figsize=(5,4))
plt.bar(counts.index,counts.values)
plt.title("Dataset Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("dataset_distribution.png",dpi=300)
plt.show()

# -------------------------
# 2. Correlation Heatmap
# -------------------------
numeric=train.select_dtypes(include="number")
corr=numeric.corr()

plt.figure(figsize=(8,6))
plt.imshow(corr,cmap="coolwarm",aspect="auto")
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation=90,fontsize=6)
plt.yticks(range(len(corr.columns)),corr.columns,fontsize=6)
plt.title("Feature Correlation")
plt.tight_layout()
plt.savefig("correlation_heatmap.png",dpi=300)
plt.show()

# -------------------------
# 3. CNN Feature PCA
# -------------------------
cnn_feat=cnn.drop(columns=["event_id"])
pca=PCA(n_components=2)
cnn_pca=pca.fit_transform(cnn_feat)

plt.figure(figsize=(6,5))
plt.scatter(cnn_pca[:,0],cnn_pca[:,1])
for i,e in enumerate(cnn["event_id"]):
    plt.text(cnn_pca[i,0],cnn_pca[i,1],str(e),fontsize=6)
plt.title("CNN Feature PCA")
plt.tight_layout()
plt.savefig("cnn_pca.png",dpi=300)
plt.show()

# -------------------------
# 4. Transformer Feature PCA
# -------------------------
tr_feat=tr.drop(columns=["event_id"])
pca2=PCA(n_components=2)
tr_pca=pca2.fit_transform(tr_feat)

plt.figure(figsize=(6,5))
plt.scatter(tr_pca[:,0],tr_pca[:,1])
for i,e in enumerate(tr["event_id"]):
    plt.text(tr_pca[i,0],tr_pca[i,1],str(e),fontsize=6)
plt.title("Transformer Feature PCA")
plt.tight_layout()
plt.savefig("transformer_pca.png",dpi=300)
plt.show()

print("Generated files:")
print("- dataset_distribution.png")
print("- correlation_heatmap.png")
print("- cnn_pca.png")
print("- transformer_pca.png")

files.download("dataset_distribution.png")
files.download("correlation_heatmap.png")
files.download("cnn_pca.png")
files.download("transformer_pca.png")
```
